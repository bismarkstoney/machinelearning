```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
hue_colors={'No':'red', 'Yes': 'black'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue="smoker", hue_order=['No', 'Yes'], palette =hue_colors)

plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker', row='time')
plt.show()
```

```python
tips.head()
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', size='size', hue='size', style='size', alpha=0.4)
plt.show()
```

```python
sns.catplot(data=tips, y='total_bill', x='time', kind='box', whis=[0,100])
```

```python
cta_order=[1,6,5,4,3,2]
sns.catplot( data=tips, x='size', kind='count', order=cta_order)
```

```python
tips.info()
```

```python
sns.catplot(x='day', y='total_bill', kind='bar', data=tips)
```

```python
from numpy import  median
sns.catplot(data=tips,x='smoker', y='total_bill', kind='point' , estimator=median, capsize=0.2)
```

```python
sns.distplot(tips['tip'])
```

```python
sns.regplot(x='total_bill', y='tip', data=tips)
```

```python
for p in sns.palettes.SEABORN_PALETTES:
    sns.set_palette(p)
    sns.distplot(tips['total_bill'])
```

```python
sns.heatmap(tips.corr(), annot=True)
```
